```python
def make_plot_1(foldername, filename, save = False, show_y = True):
    os.chdir("C:/Users/liuch/Dropbox (Personal)/MIT Pre-Doc Personal/Rahul Singh/Debiased Kernel Methods/New Code/results/401k/" + foldername)
    df = pd.read_csv(filename + ".csv")
    fig, ax = plt.subplots()

    capsize = 4
    linewidth =  1.5
    markersize = 8
    size_label = 14
    size_ticklabel = 12
    
    v = df["v"]
    mean_est = df["mean_est"]
    mean_SE = df["mean_SE"]

    plt.plot(np.linspace(25, 64, 391), effect, color = 'grey', linewidth = linewidth)
    plt.plot(v, mean_est, color = 'black', linewidth = linewidth)
    plt.errorbar(x = v, y = mean_est, yerr = mean_SE, fmt = '.', color = 'black', capsize = capsize, linewidth = linewidth, markersize = markersize)

    if show_y == True:
        plt.yticks([-15000, -10000, -5000, 0, 5000, 10000, 15000, 20000, 25000, 30000], fontsize = size_ticklabel)
        plt.ylabel("Effect on Assets", fontsize = size_label)
    else:
        plt.yticks([-15000, -10000, -5000, 0, 5000, 10000, 15000, 20000, 25000, 30000], fontsize = size_ticklabel, labels = [],)
    plt.xticks(fontsize = size_ticklabel)
    plt.xlabel("Age", fontsize = size_label)

    plt.ylim((-16000, 32000))
    plt.xlim((24, 66))

    ax.axhline(0, linestyle = '--', color = 'black')
    fig.tight_layout()
    ax.set_aspect(0.000875, adjustable = 'box')  # Makes the axes the same length

    if save == True:
        plt.savefig(filename + ".png", bbox_inches = "tight")
    plt.show()
```

```python
def CATE_fn(v):
    return v * (1+2*v)**2 * (v-1)**2 

def make_plot_2(save = False):
    os.chdir("C:/Users/liuch/Dropbox (Personal)/MIT Pre-Doc Personal/Rahul Singh/Debiased Kernel Methods/New Code/results/401k/")
    fig, ax = plt.subplots()

    linewidth =  1.5
    size_label = 14
    size_ticklabel = 12
    
    v_array_1 = np.linspace(-0.5, 0.5, 1000)
    v_array_2 = np.array([-0.25, 0, 0.25])
    plt.plot(v_array_1, CATE_fn(v_array_1), color = 'grey', linewidth = linewidth, zorder = 1)
    plt.scatter(v_array_2, CATE_fn(v_array_2), color = 'black', zorder = 2)

    plt.ylabel("$\mathrm{CATE}(\mathit{v})$", fontsize = size_label)
    plt.xlabel("$\mathit{v}$", fontsize = size_label)

    plt.xticks([-0.5, -0.25, 0, 0.25, 0.5], fontsize = size_ticklabel)
    plt.yticks(fontsize = size_ticklabel)

    plt.ylim((-0.15, 0.55))
    plt.xlim((-0.5, 0.5))

    ax.axhline(0, linestyle = '--', color = 'black')
    fig.tight_layout()
    ax.set_aspect(1/0.7, adjustable = 'box')  # Makes the axes the same length

    if save == True:
        plt.savefig("CATE_fn.png", bbox_inches = "tight")
    plt.show()
```
